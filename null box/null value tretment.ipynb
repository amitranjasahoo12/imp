# write down a code which will give me the result of null values for numerical variable and categorical varible and treat the numerical varibale null values by replacing with median and categorical variable with mode and if a varibale is a having 95% of null value drop that column?

```python
# importing libraries
import pandas as pd
import numpy as np
  
nums = {'Student Name': [ np.nan, 'Shivansh', 'Ishdeep',  
                         'Siddharth', np.nan, 'Prakhar',
                         'Yash', 'Srikar', 'Kaustubh', 
                         'Aditya',  'Manav', np.nan],
        'Roll No.': [ 18229, 18232, np.nan, 18247, 18136, 
                     np.nan, 18283, 18310, 18102, 18012,
                     18121, 18168],
       "Location": [np.NaN, np.NaN, 0, 1,1, 1, np.NaN,1, np.NaN,np.NaN, np.NaN, 0],
        "Class": [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan],
        "Sex":[0,0,np.nan,np.nan,1,1,np.nan,np.nan,1,1,np.nan,np.nan],
        'Subject ID': [204, np.nan, 201, 105, np.nan, 204,
                       101, 101, np.nan, 165, 715, np.nan],
       'Grade Point': [9, np.nan, 7, np.nan, 8, 7, 9, 10,
                       np.nan, 9, 6, 8],
        "Age": [np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan,np.nan]
       }
  
    # Create the dataframe
df = pd.DataFrame(nums)
  
    
df
```

```python
for col in df.columns: 
    if sum(df[col].isnull())/float(len(df.index)) > 0.95: 
        del df[col]
            
c = df.select_dtypes(np.number).columns
df[c]= df[c].fillna(df.median())
df=df.fillna(df.mode())
```
